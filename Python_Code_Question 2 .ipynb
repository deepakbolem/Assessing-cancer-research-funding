```python
Dataset1_index = Dataset1.set_index("Race")
Dataset1_index.head()
```

```python
Dataset1_index.reset_index()
```

```python
for c in Dataset1_index.columns:
    miss = Dataset1_index[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has No missing value(s)".format(c))
```
