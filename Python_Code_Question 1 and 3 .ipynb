```python
#cancer_research_funding_missing_values
for c in cancer_research_funding.columns:
    miss = cancer_research_funding[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has No missing value(s)".format(c))
```

```python
#cancer_cost_by_phase_missing_values
for c in cancer_cost_by_phase.columns:
    miss = cancer_cost_by_phase[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has No missing value(s)".format(c))

```

```python
# Fill missing values in cancer_research_funding
cancer_research_funding['2015 Actual'].fillna(value=cancer_research_funding['2015 Actual'].mean(), inplace=True)
cancer_research_funding['2016 Actual'].fillna(value=cancer_research_funding['2016 Actual'].mean(), inplace=True)
cancer_research_funding['2017 Actual'].fillna(value=cancer_research_funding['2017 Actual'].mean(), inplace=True)
cancer_research_funding['2018 Actual'].fillna(value=cancer_research_funding['2018 Actual'].mean(), inplace=True)
cancer_research_funding['2019 Estimate'].fillna(value=cancer_research_funding['2019 Estimate'].mean(), inplace=True)
cancer_research_funding['2020 Estimate'].fillna(value=cancer_research_funding['2020 Estimate'].mean(), inplace=True)
```

```python
# Fill missing values in cancer_cost_by_phase
cancer_cost_by_phase['Initial (in millions)'].fillna(value=cancer_cost_by_phase['Initial (in millions)'].mean(), inplace=True)
cancer_cost_by_phase['Continuing (in millions)'].fillna(value=cancer_cost_by_phase['Continuing (in millions)'].mean(), inplace=True)
cancer_cost_by_phase['End of life cancer death (in millions)'].fillna(value=cancer_cost_by_phase['Initial (in millions)'].mean(), inplace=True)
cancer_cost_by_phase['End-of-life other cause of death (in millions)'].fillna(value=cancer_cost_by_phase['End-of-life other cause of death (in millions)'].mean(), inplace=True)
cancer_cost_by_phase['Total-out-of-pocket costs (in millions)'].fillna(value=cancer_cost_by_phase['Total-out-of-pocket costs (in millions)'].mean(), inplace=True)

```

# Business Question - 3

```python
#cancer_cost_by_phase + cancer_research_funding
Merged_dataframe = pd.merge(cancer_cost_by_phase, cancer_research_funding, on = 'Cancer Site')
Merged_dataframe
```

```python
Merged_dataframe.round(2)
```

```python
#Rename columns
Merged_dataframe.rename(columns={"2015 Actual": "2015 Research(in millions)", 
                                  "2016 Actual": "2016 Research(in millions)",
                                  "2017 Actual": "2017 Research(in millions)",
                                  "2018 Actual": "2018 Research(in millions)",
                                  "2019 Estimate": "2019 Research(in millions)",
                                  "2020 Estimate": "2020 Research(in millions)"
                                 }, inplace = True)
```

```python
#funding in 2019/Total-out-of-pocket costs in 2019
Merged_dataframe["Funding per dollar out of pocket cost"] = Merged_dataframe["2019 Research(in millions)"]/ Merged_dataframe["Total-out-of-pocket costs (in millions)"]
```

```python
# plot the dataframe
Merged_dataframe.plot(x="Cancer Site", y=["Funding per dollar out of pocket cost"], kind="bar", figsize=(9, 8))
```
